**Self-assessment:**

```python
question='Choose all of the following that can be included in Jupyter notebooks?'
answers=[
    {'answer':"Formatted text",
     'correct': True,
     'feedback': "Correct."
    },
    {'answer': "Typeset mathematics",
     'correct': True,
     'feedback': "Correct."
    },
    {'answer': "Python executable code",
     'correct': True,
     'feedback': "Correct."
    },
    {'answer': "Text and graphics output from Python",
     'correct': True,
     'feedback': "Correct."
    },
    {'answer': "Live snakes via Python",
     'correct': False,
     'feedback': "I hope not."
    }
]
questions+=[ {'question':question, 'type':'many_choice', 'answers':answers} ]

display_quiz(question, answers, multiple=True)
```

```python
question='Which of these are used to create formatted text in Jupyter notebooks?'
answers=[
    {'answer':"Markdown",
     'correct': True,
     'feedback': "Correct."
    },
    {'answer': "Wiki markup",
     'correct': False,
     'feedback': "False."
    },
    {'answer': "Rich Text",
     'correct': False,
     'feedback': "False."
    },
    {'answer': "SVG",
     'correct': False,
     'feedback': "False."
    }
]
questions+=[ {'question':question, 'type':'multiple_choice', 'answers':answers} ]

display_quiz(question, answers)
```

```python
question='Which of these are used to typeset mathematiccs in Jupyter notebooks?'

answers=[
    {'answer':"LaTeX",
     'correct': True,
     'feedback': "Correct."
    },
    {'answer': "MS Equation",
     'correct': False,
     'feedback': "False."
    },
    {'answer': "SilIcoN",
     'correct': False,
     'feedback': "False."
    },
    {'answer': "AcrYliC",
     'correct': False,
     'feedback': "False."
    }
]

questions+=[ {'question':question, 'type':'multiple_choice', 'answers':answers} ]



display_quiz(question, answers, )
```

```python
question='Which of the following are used to create functional code blocks in Python?'
answers=[
    {'answer':"curly braces {}",
     'correct': False,
     'feedback': "Curly braces are used to group code in C, C++, and other languages, but not Python"
    },
    {'answer': "indentation",
     'correct': True,
     'feedback': "Correct."
    },
    {'answer': "parentheses ()",
     'correct': False,
     'feedback': "False."
    },
    {'answer': "blank lines",
     'correct': False,
     'feedback': "Blank lines can help distinguish the end of code blocks visually," +\
         " but do not affect the functional grouping"
    }
]
questions+=[ {'question':question, 'type':'multiple_choice', 'answers':answers} ]

display_quiz(question, answers)

```

```python
question='Choose all of the following that are used to create comments in Python:'
answers=[
    {'answer':"# comment",
     'correct': True,
     'feedback': "Correct. A single hash is used to create a single-line comment"
    },
    {'answer': "/* comment */",
     'correct': False,
     'feedback': "Incorrect. This is a C-style comment."
    },
    {'answer': "''' comment'''",
     'correct': True,
     'feedback': "Correct. This is used for multi-line comments."
    },
    {'answer': "//",
     'correct': False,
     'feedback': "Incorrect. This is the format of a single-line comment in C."
    }
]
questions+=[ {'question':question, 'type':'many_choice', 'answers':answers} ]





display_quiz(question, answers, multiple=True)
```
